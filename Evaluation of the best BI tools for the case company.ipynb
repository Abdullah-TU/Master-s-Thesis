```python
def m_avg(lists): 
    
    """
    Calculate and return the average values of a list. 
    
    : param lists: list, the parameter lists take values in a list, calculate them, and return the average value.   
    : return averge: float, the average values in a new list.
    
    """
    add = [[sum(i)/len(i)] for i in lists]
    averge = []
    for i in add:
        for j in i:
            averge.append(float(j))            
    return averge
```

```python
def dic_create(list1, list2): 
    
    """
    dic_create() function takes two parameters. list1 as keys and list2 as values and 
    returns a dictionary of those keys and values. 
    
    : param list1: list, the keys of the dictionary
    : param list2: list, the values of the dictionary
    : return: a dictionary, a dictionary using keys and values.
    
    """        
    dictiony=dict(zip(keys, list2))
    return dictiony
```

```python
def diction_value(dictiony):  
    
    """ 
    diction_value() function takes one parameter. Its a dictionary type parameter. 
    diction_value() function returns the sum of the values of the dictionary.
    
    : param dictiony: dictionary, to find out sum of the values of the dictionary.
    : return: float, returns the sum of the values of the dictionary.
    
    """        
    dict_value= sum(dictiony.values())
    return dict_value
```

```python
def another_dict(srt, dicts):
    
    """
    another_dict() function takes two parameters. One is a string type parameter, and another is the dictionary.  
    another_dict() function returns the sum of the values of the dicitonary.
    
    : param srt: string, as a key of the dictionary.    
    : param dicts: dictionary, values of the dictionary are summed and returned the sum of the values.
    : return: float, the sum of the values of the dictionary parameter.
    
    """    
    value=sum(dicts.values())   
    return value
```

```python
def final(tsil1, tsil2, tsil3):
    
    """
    finalfunction() takes three parameters. tsil1 and tsil2 are list, tsil3 is string. Here the special formula is used to 
    find out the weighted score(new_dic2). After the calculation finalfunction() function returns float, a total weighted 
    score. Multiple functions are used, and several calculations have been conducted to find out the final expeted result. 
        
    : param tsil1: list, the main list passed to calculate.    
    : param tsil2: list, the keys of the dictionary.
    : param tsil3: string, passed the name of the BI tool from the list of them.    
    : return: float, the total weighted score.
    
    """
        
    dic1 = dic_create(tsil2, m_avg(tsil1))
    dic2=diction_value(dic1)
    new_dic2 = [[(dic2/5)*dic1['Efficacy 1']],[(dic2/5)*dic1['Efficacy 2']],[(dic2/5)*dic1['Efficacy 3']],
                [(dic2/5)*dic1['Efficacy 4']], [(dic2/5)*dic1['Efficacy 5']],[(dic2/5)*dic1['Efficacy 6']],
                [(dic2/5.)*dic1['Efficacy 7']],[(dic2/5)*dic1['Efficacy 8']],[(dic2/5)*dic1['Efficacy 9']], 
                [(dic2/5)*dic1['Efficacy 10']]]
    
    dic3 = dic_create(dic2, m_avg(new_dic2))
    BIdic=another_dict(tsil3, dic3)
    finalBI=round(BIdic,2)
    return finalBI
```
