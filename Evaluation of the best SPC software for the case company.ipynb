```python
def m_avg(lists): 
    
    """
    Calculate and return the average values of a list. 
    
    : param lists: list, the parameter lists take values in a list, calculate them, and return the average value.   
    : return averge: float, the average values in a new list.
    
    """
    add = [[sum(i)/len(i)] for i in lists]
    averge = []
    for i in add:
        for j in i:
            averge.append(float(j))            
    return averge
```

```python
def dic_create(list1, list2): 
    
    """
    dic_create() function takes two parameters. list1 as keys and list2 as values and 
    returns a dictionary of those keys and values. 
    
    : param list1: list, the keys of the dictionary
    : param list2: list, the values of the dictionary
    : return: a dictionary, a dictionary using keys and values.
    
    """        
    dictiony=dict(zip(keys, list2))
    return dictiony
```

```python
def diction_value(dictiony):  
    
    """ 
    diction_value() function takes one parameter. Its a dictionary type parameter. 
    diction_value() function returns the sum of the values of the dictionary.
    
    : param dictiony: dictionary, to find out sum of the values of the dictionary.
    : return: float, returns the sum of the values of the dictionary.
    
    """       
    dict_value= sum(dictiony.values())
    return dict_value
```

```python
def another_dict(srt, dicts):
    
    """
    another_dict() function takes two parameters. One is a string type parameter, and another is the dictionary.  
    another_dict() function returns the sum of the values of the dicitonary.
    
    : param srt: string, as a key of the dictionary.    
    : param dicts: dictionary, values of the dictionary are summed and returned the sum of the values.
    : return: float, the sum of the values of the dictionary parameter.
    
    """    
    value=sum(dicts.values())   
    return value
```

```python
def final(tsil1, tsil2, tsil3):
    
    """
    finalfunction() takes three parameters. tsil1 and tsil2 are list, tsil3 is string. Here the special formula is used to 
    find out the weighted score(new_dic2). After the calculation finalfunction() function returns float, a total weighted 
    score. Multiple functions are used, and several calculations have been conducted to find out the final expeted result. 
        
    : param tsil1: list, the main list passed to calculate.    
    : param tsil2: list, the keys of the dictionary.
    :param tsil3: string, passed name of the SPC tool from the list of them.  
    : return: float, the total weighted score.
    
    """

        
    dic1 = dic_create(tsil2, m_avg(tsil1))
    dic2=diction_value(dic1)
    new_dic2 = [[dic2/5.0*dic1['Efficacy 1']],[dic2/5.0*dic1['Efficacy 2']],[dic2/5*dic1['Efficacy 3']],
                [dic2/5*dic1['Efficacy 4']], [dic2/5*dic1['Efficacy 5']],[dic2/5.0*dic1['Efficacy 6']], 
                [dic2/5.0*dic1['Efficacy 7']]]
    dic3 = dic_create(dic2, m_avg(new_dic2))
    BIdic=another_dict(tsil3, dic3)
    finalBI=round(BIdic,2)
    return finalBI
```

# Best 13 SPC software for the case company

```python
""" 
SPCSoft_dictionary: a dictionary, takes keys and values.
SPCSoft: list, passing lists of the name of the SPCSoft tools.
SPCSoft[0], first element of SPCSoft list, as key and SPCSoft[1], SPCSoft[2], .... SPCSoft[12] shows the name of SPCSoft 
tools respectively.

For example, WinSPC: total weighter value of the specific SPCSoft tool. WinSPC, Enact, ..., QDM_SPC shows the total 
weighted values respectively.

"""

SPCSoft_dictionary={SPCSoft[0]:WinSPC, SPCSoft[1]:Predisys, SPCSoft[2]:Enact , SPCSoft[3]:SPC_for_Excel, SPCSoft[4]:DataLyzer ,
SPCSoft[5]:IQMS_MES, SPCSoft[6]:Intuitive_ERP , SPCSoft[7]:iPASP_Cloud_SPC, SPCSoft[8]:RealTime, SPCSoft[9]:Synergy, 
SPCSoft[10]:SQCpack, SPCSoft[11]:ProFicient, SPCSoft[12]:QDM_SPC }

SPCSoft_dictionary
```

# The best of the best for the case company

```python
""" 
best: the sorted() function, sorted the key (name of the SPC software) which contains highest weighted values out of 
all 13 tools.

"""

best = sorted(SPCSoft_dictionary, key=SPCSoft_dictionary.get, reverse=True)[:1]

for i in best:
    print(i)
```

```python
""" 
import: importing visualization libraries.
fig: represents the figure size and face color of the plot.
softdic: represents the dictionary which contains the dictionary with the name of the SPC software and their total weighted
values.
main_plot: represents the values taken from d for the x-axis and the y-axis.

Finally, the best SPC software for the case company was sorted out and visualized using different colors.

"""

import matplotlib.pyplot as plt
import seaborn 
import numpy
import matplotlib.image as image

fig = plt.figure(figsize=(15,9),facecolor='#004d66')

softdic = SPCSoft_dictionary
main_plot = seaborn.barplot(x=[x for x in softdic.keys()],y=[x for x in softdic.values()]) 

for i in main_plot.patches:
    if i.get_height() > 210:
        i.set_color('#a834a8')        
    else:
        i.set_color('#42b7bd')
        
plt.suptitle('Selection the best SPC Software for the case company', fontsize=20,color='w',weight="bold")
plt.xticks(rotation=55, ha='right',color='w', fontsize=12)
plt.yticks(color='w', fontsize=12)
plt.tick_params(color='w',  width=3,length=6)
plt.bar_label(main_plot.containers[0])
plt.xlabel('Best 13 SPC Software for the case company', fontsize=20,color='w',weight="bold")
plt.ylabel('Comprehensive Score', fontsize=20,color='w',weight="bold")
plt.rcParams['axes.facecolor'] = '#e6e6ff'

fig.text(0.86, 0.96, ' Best 1 ', color='w' , size=11, weight="bold", bbox=dict(facecolor='#a834a8' , edgecolor='w', boxstyle='round, pad=0.6'))

img = image.imread("pic1.png")
plt.figimage(img, 1.66, 0.96, zorder=1, alpha=0.1)

plt.tight_layout()

plt.savefig('verticalbarspc1.png')
plt.show()
```
